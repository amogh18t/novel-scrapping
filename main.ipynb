```python
# import the required libraries
import requests
from bs4 import BeautifulSoup

# define the URL of the website
url = 'https://www.wuxiaworld.com/novels'

# send a GET request to the website
response = requests.get(url)

# parse the HTML content of the website
soup = BeautifulSoup(response.content, 'html.parser')

# extract the summary of the novel
novel_summaries = soup.find_all('p', class_='summary')

# extract the text from the summaries
novel_summaries_text = []
for summary in novel_summaries:
    novel_summaries_text.append(summary.get_text())

# print the summaries
print(novel_summaries_text)

# output the summaries to a file
with open('novel_summaries.txt', 'w') as f:
    for summary in novel_summaries_text:
        f.write(summary + '\n')

# output the summaries to the console
for summary in novel_summaries_text:
    print(summary)


# Find the summaries with the help of LLMs after extracting all of the text from every novel_summaries

# Use the Hugging Face Transformers library to load the GPT-2 model
    
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the GPT-2 model and tokenizer
model = GPT2LMHeadModel.from_pretrained('gpt2')
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

# Set the model to evaluation mode
model.eval()

# Define a function to generate summaries
def generate_summary(text):
    # Tokenize the input text
    input_ids = tokenizer.encode(text, return_tensors='pt')
    
    # Generate the output summary
    output = model.generate(input_ids, max_length=100, num_return_sequences=1, no_repeat_ngram_size=2, early_stopping=True)
    
    # Decode the output summary
    summary = tokenizer.decode(output[0], skip_special_tokens=True)
    
    return summary

# Generate summaries for each novel
for summary in novel_summaries_text:
    generated_summary = generate_summary(summary)
    print(generated_summary)

# output the generated summaries to a file
with open('generated_summaries.txt', 'w') as f:
    for summary in novel_summaries_text:
        f.write(summary + '\n')
        generated_summary = generate_summary(summary)
        f.write(generated_summary + '\n')

# output the generated summaries to the console
for summary in novel_summaries_text:
    print(summary)
    generated_summary = generate_summary(summary)
    print(generated_summary)


```

```python
#  first web scrawl the site "https://wuxiaworld.com/novels" and find all novel and make a summary for one novel using all 
# of its chapters and then do it for every scraweled website and write this code using the help of LLMs on the device using mistral 7B 
# and also amke UI for it with help of streamlit and fast-api and the summary should be generated using the total text of all chapters also 
# with timeline of important events and with help of NER find the Main Character and related characters name and relation with MC
# and the summary will be made using LLMs from chapters_content
# and instead of inputing text in input box input website there and get the summary
# also add the NER to find the main character and related characters name and relation with MC
# also add the timeline of important events

# import the required libraries
import requests
from bs4 import BeautifulSoup
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import streamlit as st
import fastapi
import uvicorn
import spacy
from spacy import displacy

# define the URL of the website
url = 'https://wuxiaworld.com/novels'

# send a GET request to the website
response = requests.get(url)

# parse the HTML content of the website
soup = BeautifulSoup(response.content, 'html.parser')

# extract the summary of the novel
novel_summaries = soup.find_all('div', class_='summary')

# extract the text from the summaries
novel_summaries_text = [summary.getText().strip() for summary in novel_summaries]

# print the summaries
for summary in novel_summaries_text:
    print(summary)

# output the summaries to a file
with open('novel_summaries.txt', 'w') as f:
    for summary in novel_summaries_text:
        f.write(summary + '\n')

# output the summaries to the console
for summary in novel_summaries_text:
    print(summary)

# Use the Hugging Face Transformers library to load the GPT-2 model
model = GPT2LMHeadModel.from_pretrained('gpt2')
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model.eval()

# Define a function to generate summaries
def generate_summary(text):
    input_ids = tokenizer.encode(text, return_tensors='pt')
    output = model.generate(input_ids, max_length=100, num_return_sequences=1, no_repeat_ngram_size=2, early_stopping=True)
    summary = tokenizer.decode(output[0], skip_special_tokens=True)
    return summary

# Generate summaries for each novel
generated_summaries = []
for summary in novel_summaries_text:
    generated_summary = generate_summary(summary)
    generated_summaries.append(generated_summary)

# output the generated summaries to a file
with open('generated_summaries.txt', 'w') as f:
    for summary in generated_summaries:
        f.write(summary + '\n')

# output the generated summaries to the console
for summary in generated_summaries:
    print(summary)

#uvicorn and fast-api were not used
    
# Streamlit UI
st.title('Web Novel Summarizer')

# Add a text input box
text = st.text_input("Enter the text of the web novel:", value="")
if text:
    generated_summary = generate_summary(text)
    st.write("Generated Summary:")
    st.write(generated_summary)

# FastAPI UI
app = fastapi.FastAPI()

@app.get("/summarize")
async def summarize(text: str):
    generated_summary = generate_summary(text)
    return {"summary": generated_summary}

if __name__ == "__main__":
    uvicorn.run(app, host="0.0.0.0", port=8000)

# run the FastAPI server using the command
    
# uvicorn main:app --reload
    
# open the browser and go to http://localhost:8000/summarize?text=your_text_here
# to get the generated summary
    
# NER using spacy
nlp = spacy.load("en_core_web_sm")

# Add a text input box for NER
ner_text = st.text_input("Enter the text for NER:", value="")
if ner_text:
    doc = nlp(ner_text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    st.write("Named Entities:")
    st.write(entities)

# Add a text input box for timeline
timeline_text = st.text_input("Enter the text for timeline:", value="")
if timeline_text:
    doc = nlp(timeline_text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    st.write("Timeline:")
    st.write(entities)
    displacy.render(doc, style="ent", jupyter=True)

# Add a text input box for character
character_text = st.text_input("Enter the text for character:", value="")
if character_text:
    doc = nlp(character_text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    st.write("Character:")
    st.write(entities)
    displacy.render(doc, style="ent", jupyter=True) 



```
